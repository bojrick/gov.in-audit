```python
import pandas as pd
import json

df = pd.read_csv('data.csv')
urls = df['link'].tolist()

with open('urls.json', 'w') as json_file:
    json.dump(urls, json_file)

print("URLs have been extracted to urls.json")
```

```python
command = "lighthouse-batch -s urls.json"
process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
stdout, stderr = process.communicate()

# Check for errors
if process.returncode == 0:
    print("Lighthouse reports generated successfully.")
else:
    print(f"Error generating reports: {stderr.decode()}")
```

```python
import json
import subprocess
import socket

# Load URLs from the existing JSON file
with open('urls.json', 'r') as json_file:
    data = json.load(json_file)

# Check if the data is a list or a dict
if isinstance(data, dict):
    urls = data.get("urls", [])
elif isinstance(data, list):
    urls = data  # Directly assign the list of URLs
else:
    print("Unexpected JSON format.")
    urls = []

def check_dns(url):
    try:
        # Extract hostname from URL
        hostname = url.split("//")[-1].split("/")[0]
        ip_address = socket.gethostbyname(hostname)
        print(f"DNS resolved for {url}: {ip_address}")
        return True
    except socket.gaierror:
        print(f"DNS resolution failed for: {url}")
        return False

```

```python
# Check DNS for each URL
valid_urls = [url for url in urls if check_dns(url)]

if valid_urls:
    # Create a new JSON file with only valid URLs
    valid_urls_json = {"urls": valid_urls}
    with open('valid_urls.json', 'w') as json_file:
        json.dump(valid_urls_json, json_file)
```

```python
# Run lighthouse-batch command with valid URLs
command = "lighthouse-batch -s valid_urls.json --verbose"
process = subprocess.Popen(command, shell=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)

# Wait for the process to complete and capture output
stdout, stderr = process.communicate()

# Check for errors
if process.returncode == 0:
    print("Lighthouse reports generated successfully.")
else:
    # Decode and print the error message if the command failed
    error_message = stderr.decode().strip()
    print(f"Error generating reports: {error_message}")
```
